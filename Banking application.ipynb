```python
import sqlite3 as s
import sqlalchemy as sl
import pandas as pd
class Database:
    def createdatabase(self):
        self.con1=s.connect('data1.sqlite')
        self.con=self.con1.cursor()
        self.query='create table bank9(acc_no number(10),first_name varchar2(20),last_name varchar2(20),adhaar number(12),dob date,address varchar2(20),balance number(20),primary key(acc_no))'
        self.con.execute(self.query)
    def insert(self,d):
        statement='insert into bank9(acc_no,first_name,last_name,adhaar,dob,address,balance) values(?,?,?,?,?,?,?)'
        self.con.execute(statement,d)
        self.con1.commit()
    def view(self):
        data=self.con.execute('select * from bank9')
        info=data.fetchall()
        structinfo=pd.DataFrame(info,columns=['acc_no','first_name','last_name','adhaar','dob','address','balance'])
        print(structinfo)
db=Database()
class Createuser(Database):
    def __init__(self):
        self.acc_no=1111111111
        global db
    def createaccount(self):
        first_name=input('enter first name:')
        last_name=input('enter last name:')
        adhar_no=int(input('enter adhar number:'))
        dob=input('enter dob:')
        address=input('enter address:')
        self.acc_no=self.acc_no+1
        d=(self.acc_no,first_name,last_name,adhar_no,dob,address,0.0)
        db.insert(d)
        db.view()
if __name__=='__main__':
    c=Createuser()
    db.createdatabase()
    print("<><><><><> BANKING APPLICATION <><><><><>")
    while True:
        print('__________________________________________')
        print('...........MENU..............')
        print('1-> create new account')
        print('2->login account')
        user_chioce=int(input('enter option:'))
        if user_chioce==1:
            c.createaccount()
        elif user_chioce==2:
            userid=int(input('enter user id:'))
            password=input('enter password:')
            x=db.con.execute('select acc_no,dob from bank9 where acc_no=? and dob=?',(userid,password,))
            y=x.fetchall()
            if len(y)!=0:
                print('...........LOGIN MENU..............')
                print('1->deposit')
                print('2->withdraw')
                print('3->checkbalance')
                print('4->exit')
                uc=int(input('enter login menu option:'))
                if uc!=4:
                    bal=db.con.execute('select balance from bank9 where acc_no=?',(userid,))
                    cbal=bal.fetchall()
                    ucbal=cbal[0]
                if uc==1:
                    amount=int(input('enter amount of deposit:'))
                    newbal=ucbal[0]+amount
                    db.con.execute('update bank9 set balance=? where acc_no=?',(newbal,userid))
                    print('your amount is deposited successfully...!')
                elif uc==2:
                    amount=int(input('enter amount of withdraw:'))
                    if ucbal[0]<amount:
                        print('insufficient balance..')
                    else:
                        newbal=ucbal[0]-amount
                        db.con.execute('update bank9 set balance=? where acc_no=?',(newbal,userid))
                        print('debited rupees:',amount)
                elif uc==3:
                    print('your clear balance is:')
                    print(ucbal[0])
                elif uc==4:
                     break
                else:
                    print('invalid choice')
            else:
                print('invalid username or password')
        else:
            print('invalid choice')
    print('exited from application')     
```
